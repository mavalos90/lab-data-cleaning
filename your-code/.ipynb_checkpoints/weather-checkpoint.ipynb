## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../weather-raw.csv", sep=',')
data.head()
```

```python
# pivot table to group
# dropna = False to preserve NaN columns
df = pd.pivot_table(data, values=['d1', 'd2', 'd3', 'd4', 'd5', 'd6',
       'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd16',
       'd17', 'd18', 'd19', 'd20', 'd21', 'd22', 'd23', 'd24', 'd25', 'd26',
       'd27', 'd28', 'd29', 'd30', 'd31'], 
        index=['year', 'month', 'element'],
        dropna=False) 

df.head(10)
```

```python
#sort the columns
df = df[['d1', 'd2', 'd3', 'd4', 'd5', 'd6',
       'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd16',
       'd17', 'd18', 'd19', 'd20', 'd21', 'd22', 'd23', 'd24', 'd25', 'd26',
       'd27', 'd28', 'd29', 'd30', 'd31']]

df.head(10)
```

```python
df['mean'] = round(df.transpose().mean(),2)
df['std'] = round(df.transpose().std(),2)
df['max_value'] = round(df.transpose().max(),2)
df['min_value'] = round(df.transpose().min(),2)

resume = df

resume = resume.drop(['d1', 'd2', 'd3', 'd4', 'd5', 'd6',
       'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd16',
       'd17', 'd18', 'd19', 'd20', 'd21', 'd22', 'd23', 'd24', 'd25', 'd26',
       'd27', 'd28', 'd29', 'd30', 'd31'], axis=1)

resume.head(10)
```
